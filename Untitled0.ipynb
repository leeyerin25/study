```python

import pandas as pd
import os
from pycaret.classification import *
from sklearn.preprocessing import LabelEncoder
```

```python
#train 데이터 불러오기
train = pd.read_csv('train.csv')
# test 데이터 불러오기
test = pd.read_csv('test.csv')
#sample_submission 불러오기
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
def handle_na(data):
    temp = data.copy() #원본사용보다는 복사본 사용하기 -> 변형 과정 검토차ㅕㅓㅑ
    temp['Gender'] = temp['Gender'].replace('Fe Male','Female')
    return temp
```

```python
train = handle_na(train)
```

```python
clf = setup(data = train,
            target = 'ProdTaken',
            train_size = 0.8,
            session_id = 201,
            fold_shuffle = True,
            numeric_features = ['MonthlyIncome', 'DurationOfPitch', 'NumberOfPersonVisiting', 'NumberOfFollowups', 'NumberOfTrips', 'NumberOfChildrenVisiting'],
            categorical_features = ['TypeofContact', 'CityTier', 'Gender', 'Occupation', 'ProductPitched', 'MaritalStatus', 'Passport', 'OwnCar', 'Designation'],
            # ordinal_features = ['CityTier', 'PitchSatisfactionScore'],
            numeric_imputation = 'mean',
            ignore_features = ['id'],
            normalize = True, normalize_method = 'robust',
            transformation = True,
            ignore_low_variance = True,
            combine_rare_levels = True, rare_level_threshold = 0.05,
            remove_outliers = True, outliers_threshold = 0.05,
            remove_multicollinearity = True, multicollinearity_threshold = 0.90,
            create_clusters = True, cluster_iter = 10,
            silent = True,
            feature_selection = True, feature_selection_threshold = 0.80,
            verbose = False
            )
```

```python
top5 = compare_models(sort='Accuracy' , n_select = 5)
```

```python
#소프트보팅
blender_top5 = blend_models(estimator_list=top5, method='auto')
```

```python
stacker_top5 = stack_models(top5, meta_model=None)
```

```python
pred_holdout = predict_model(stacker_top5)
```

```python
final_model = finalize_model(stacker_top5)
```

```python
pred_unseen = predict_model(final_model, test)
```
